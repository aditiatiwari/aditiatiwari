```python
import yfinance as yf
import pandas as pd
```

```python
def calculate_monthly_returns(tickers, start_date, end_date):
    returns_dict = {}
    for ticker in tickers:
        stock_data = yf.download(ticker, start=start_date, end=end_date)['Adj Close']
        
        monthly_returns = stock_data.resample('M').ffill().pct_change().dropna()
        
        returns_dict[ticker] = monthly_returns
    return returns_dict
```

```python
bank_tickers = ['TD', 'JPM', 'BAC', 'C', 'MS', 'USB', 'WFC', 'PNC', 'FITB', 'RF', 'MTB', 'CFG', 'TFC', 'GS', 'KEY']
start_date = '2020-12-31'
end_date = '2024-12-31'

monthly_returns_dict = calculate_monthly_returns(bank_tickers, start_date, end_date)

HighCategory = ['TD', 'JPM', 'BAC', 'C', 'MS']
MediumCategory = ['USB', 'WFC', 'PNC', 'FITB', 'RF']
LowCategory = ['MTB', 'CFG', 'TFC', 'GS', 'KEY']

HighCategory_df = pd.DataFrame({ticker: returns for ticker, returns in monthly_returns_dict.items() if ticker in HighCategory})
MediumCategory_df = pd.DataFrame({ticker: returns for ticker, returns in monthly_returns_dict.items() if ticker in MediumCategory})
LowCategory_df = pd.DataFrame({ticker: returns for ticker, returns in monthly_returns_dict.items() if ticker in LowCategory})

print(" High Category:")
print(HighCategory_df)

print("\nMedium Category:")
print(MediumCategory_df)

print("\nLow Category:")
print(LowCategory_df)
```

```python
HighCategory_df.to_excel('HighC_monthly_returns.xlsx', index=True)
MediumCategory_df.to_excel('MediumC_monthly_returns.xlsx', index=True)
LowCategory_df.to_excel('Low_monthlyC_returns.xlsx', index=True)

print("Data exported to Excel files successfully.")
```

```python
import pandas as pd
import plotly.express as px
import seaborn as sns

# Replace 'file_path.xlsx' with the actual path to your Excel file
file_path = r"C:\Users\aditi\Downloads\Pandas DF.xlsx"

# Read the Excel file into a Pandas DataFrame
df = pd.read_excel(file_path, engine='openpyxl')

# Create a line plot for HCB with IRC
fig = px.line(df, x='Date', y=['High Ranged D/E Group Companies Average Monthly Returns', '10-Year Real Interest Rate Monthly Changes'], title='Monthly Returns Of Companies Divided In 3 Groups vs. IRC Over Time')

# Add lines for MCB with IRC and LCB with IRC
fig.add_scatter(x=df['Date'], y=df['Medium Ranged D/E Group Companies Average Monthly Returns'], mode='lines', name='Medium Ranged D/E Group Companies Average Monthly Returns')
fig.add_scatter(x=df['Date'], y=df['Low Ranged D/E Group Companies Average Monthly Returns'], mode='lines', name='Low Ranged D/E Group Companies Average Monthly Returns')

# Update x-axis to display labels on a monthly basis
fig.update_xaxes(dtick='M1', tickformat='%b\n%Y', title_text="Date")

# Update y-axis label
fig.update_yaxes(title_text="Changes in Returns & Interest Rates in %")

# Update legend title
fig.update_layout(legend_title_text="Legend")

# Show the plot
fig.show()
```
